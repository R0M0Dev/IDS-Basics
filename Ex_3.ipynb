```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Datasets/Emp_visu.csv')
```

01. Draw a bar chart with Team and its count (use different colors for each team) 

```python
team_count = df['Team'].value_counts()
plt.bar(team_count.index,team_count.values,color=['red','blue','green','gold','black','grey','purple'])
plt.xticks(rotation=90)
plt.show()
```

02. Draw a comparative bar chart for Salary and New_Salary against each person (first 15 persons)

```python
plt.bar(df['First Name'],df['Salary'],width=0.5,label='salary')
plt.bar(df['First Name'],df['New_Salary'],color='red',width=0.3,align='edge',label='new salary')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

03. Draw a horizontal bar chart for Team and Salary

```python
plt.barh(df['Team'],df['Salary'])
plt.show()
```

04. Draw a stacked bar chart for Salary and New_price against the person (first 10 persons)

```python
temp_df = df.head(10)

plt.bar(temp_df['First Name'],temp_df['New_Salary'])
plt.bar(temp_df['First Name'],temp_df['Salary'],color='red')
plt.xticks(rotation=45)
plt.show()
```

05. Draw a pie chart with Gender and its count

```python
val_count = df['Gender'].value_counts()
plt.pie(val_count.values,labels=val_count.index,autopct='%1.1f%%',colors=['blue','red'])
plt.show()
```

06. Draw the dot plot between person and experience (first 15 persons)

```python
df_temp = df.head(15)
plt.scatter(df_temp["First Name"],df_temp["Experience"])
plt.xticks(rotation=90)
plt.show()
```

07. Draw the line plot between age and experience. Observe the trend line. 

```python
plt.plot(df['Age'],df['Experience'],marker='o')
plt.show()
```

08. Draw the scatter plot between Salary and New_Salary. Observe the correlation

```python
plt.scatter(df['Salary'],df['New_Salary'])
```

09. Draw the scatter plot between Age and Incentive. Observe the correlation

```python
plt.scatter(df['Age'],df['Incentive'])
plt.show()
```

10. Draw the box plot to show the statistical summary of Age column and verify with describe()

```python
plt.boxplot(df['Age'])
plt.show()
df['Age'].describe()
```

11. Draw the histogram plot for Experience column

```python
plt.hist(df['Experience'])
plt.show()
```

12. Draw the histogram plot for Experience column with bin value and PDF

```python
plt.hist(df['Experience'],bins=10,density=True)
```
