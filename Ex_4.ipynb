```python
import pandas as pd
import matplotlib.pyplot as plt
df =pd.read_csv('Datasets/iris_EDA.csv')
```

1. Remove the irrelevant column "Color" (inplace=True)

```python
df.drop('Color',axis=1,inplace=True)
df
```

2. Remove the duplicate rows and analyze

3. Rename the column class to Category


```python
df.rename(columns={"class":"Category"},inplace=True)
df
```

4. Calculate the central tendency measures for "Score"

```python
mean = df['Score'].mean()
median = df['Score'].median()
mode = df['Score'].mode()
print('mean',mean)
print('mean',median)
print('mode',mode[0])
```

5. Calculate the variability measures for "Score"

```python
df['Score'].describe()
```

6. Calculate the IQR using quantile for "Score"

```python
IQR = df['Score'].quantile(0.75) - df['Score'].quantile(0.25)
print(IQR)
```

7. Calculate the z-score for "Score"

```python
mean = df['Score'].mean()
std = df['Score'].std()
df['z-score'] = (df['Score']-mean)/std
df
```

10. Replace the nan value in petalwidth column with previous value, next value, linear interpolation and central tendency measures.

```python
df['petalwidth_pre'] = df['petalwidth'].ffill()
df['petalwidth_next'] = df['petalwidth'].bfill()
df['petalwidth_intpolate'] = df['petalwidth'].interpolate()
df['petalwidth_mean'] = df['petalwidth'].fillna(df['petalwidth'].mean())
df['petalwidth_median'] = df['petalwidth'].fillna(df['petalwidth'].median())
df['petalwidth_mode'] = df['petalwidth'].fillna(df['petalwidth'].mode()[0])
df
```

12. Remove the outliers using IQR by recalculating IQR in updated "Score" and analyse with boxplot

```python
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3-Q1
df_no_outlier = df[(df['Score'] >= Q1 - 1.5 *IQR) & (df['Score'] <= Q1 + 1.5 *IQR )]
df_no_outlier
```

13. Remove the outliers using z-score by recalculating z-score in updated "Score" and analyse with boxplot


```python
threshold = 3
new_df = df[(df['z-score']<threshold)]
plt.boxplot(new_df['Score'].dropna())
plt.show()
```
